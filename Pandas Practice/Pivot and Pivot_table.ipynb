```python
import pandas as pd
df = pd.read_excel("weather.xlsx")
df
```

```python
# Pivote table is used to summarize and aggregate data inside dataframe

# if you wanted to display "Day" as index follow as shown
df.pivot(index="Day", columns="City")
```

```python
# if you wanted to display only temprature the add one more argument as shown
df.pivot(index="Day",columns="City",values="Temperature")
```

```python
# we also have "aggfunc=" function inside ".pivot" here we can find sum,mean,count, etc
df.pivot_table(index="City", columns="Day", values="Temperature", aggfunc="mean")
```
