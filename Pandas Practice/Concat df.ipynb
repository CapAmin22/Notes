```python
import pandas as pd
india_weather = pd.DataFrame({
    "City":['Mumbai','Delhi','Banglore'],
    "temprature":[32,45,30],
    "Humidity": [80,60,78]})
india_weather
```

```python
us_weather = pd.DataFrame({
    "City":['New York','Chicago','Orlando'],
    "temprature":[21,14,35],
    "Humidity": [68,65,75]})
us_weather
```

```python
# Concatenating the both DF
df = pd.concat([india_weather,us_weather])
df
```

```python
# correcting the index numbers which is not correct in previous Concatenatinated DF\
df = pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
# using keys to specify the table
df = pd.concat([india_weather,us_weather], keys = ["india","us"])
df          
```

```python
#extracting a single table using ".loc[]"
df.loc["us"]
```

```python
# If we have shuffled data we can use "index=[1,0]" 
# to alter the position of first variables data.
# here city variable is shuffled
india_weather = pd.DataFrame({
    "City":['Delhi','Mumbai','Banglore'],
    "temprature":[32,45,30],
    "Humidity": [80,60,78],}, index=[1,0,2])
india_weather
```

```python
# if you want to add one more variable in DF then follow as shown
events = pd.DataFrame({
    'Events': ['Humid','Dry','Humid'],
})
events
```

```python
pd.concat([india_weather,events], axis=1)
```

```python
# we can also concat the data with "pd.series" library
df = pd.Series(['Humid','Dry','Humid'], name = 'Events')
event = pd.concat([india_weather,df], axis=1)
event
```
