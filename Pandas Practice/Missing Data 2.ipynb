```python
import pandas as pd
import numpy as np
data = pd.read_excel("2)weather_data.xlsx")
df = pd.DataFrame(data)
df
```

```python
# if you wanted to replace value you can do it as follows
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
# if you wanted to replace multiple variable data the follow as below
new_df = df.replace({
    'temprature': -99999,
    'windspeed': -99999,
    'event': 0},np.NaN)
new_df
```

```python
# mapingof valu you want to replace.
new_df = df.replace({
    -99999: np.NaN,
    'No Event':'Sunny'},np.NaN)
new_df
```

```python
# if you want to extract the alphabet from a cell/variable we can do it as follow.
new_df = df.replace({
    'temprature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'},'',regex=True)
new_df
```
