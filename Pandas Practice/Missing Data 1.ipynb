```python
# tpics we covering
# fillna to fill missing values using different ways. 
# interpolate to make a guess on missing values using interpolation
# dropna to drop rows with missing values.
```

```python
import pandas as pd
# if you want to convert Day value into Date/Timestamp follow as below
weather_data = pd.read_excel("C:\\Users\\HP\\Pandas Practice\\weather_data.xlsx", parse_dates=["Day"])
weather_data
```

```python
type(weather_data.Day[0])
```

```python
# if you want to set Day index follow as below
weather_data.set_index('Day', inplace=True)
weather_data
```

```python
# if you dont want to modify the previous data-fram permanuntly 
# but wanted to replace the Value of NaN temporarily then follow as bellow. 
new_weather_data = weather_data.fillna(0)
new_weather_data
```

```python
# if you want to replace multiple column NaN-data then follow as bellow
new_weather_data = weather_data.fillna ({
    'Day': 0,
    'temprature' : 0,
    'windspeed' : 0,
    'event' : 'No event'
})
new_weather_data
```

```python
# if you want to carry forwared the previous cell data then will use 
# ".fillna(method='ffill')" as below.
new_weather_data = weather_data.fillna(method='ffill')
new_weather_data
```

```python
# if you want to fill the next cell data in the present cell then will use 
# ".fillna(method='bfill')" as below.
# by using limit=1 we can limit the cell 
new_weather_data = weather_data.fillna(method='bfill',limit=1)
new_weather_data
```

```python
# Linear interpolation is a method used to estimate values 
# that fall between two known data points.".interpolate()"
new_weather_data = weather_data.interpolate()
new_weather_data
```

```python
# DataFrame will have the missing values replaced 
# with interpolated values based on the time index.".interpolate(method='time')"
new_weather_data = weather_data.interpolate(method = 'time')
new_weather_data
```

```python
# The new_weather_data DataFrame will only contain 
# the rows with no missing values in any of the columns.".dropna()"
new_weather_data = weather_data.dropna()
new_weather_data
```

```python
#  new_weather_data DataFrame will exclude rows that 
# have all missing values in all columns.".dropna(how='all')"
new_weather_data = weather_data.dropna(how='all')
new_weather_data
```

```python
# This means that you want to keep only those 
# rows that have at least 1 non-missing value. ".dropna(thresh=1)"
# we can increas the number of non-missing value according our requirment.
new_weather_data = weather_data.dropna(thresh=1)
new_weather_data
```
